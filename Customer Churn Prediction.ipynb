```python
df1= df[df.TotalCharges!=' ']
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python
tensure_column_no = df1[df1.Churn == 'No'].tenure
tensure_column_yes = df1[df1.Churn == 'Yes'].tenure
plt.xlabel('tenure')
plt.ylabel('Number of Customer')
plt.hist([tensure_column_yes, tensure_column_no], color = ['green', 'red'], label  =['Yes', 'No'])
plt.legend()
```

```python
def print_unique():
    for column in df1:
        if df1[column].dtypes == 'object':
            print(f'{column} : {df1[column].unique()}')
```

```python
print_unique()
```

```python
df1.replace('No internet service', 'No', inplace=True)
df1.replace('No phone service', 'No', inplace=True)
```

```python
print_unique()
```

```python
yes_no_columns = ['Partner', 'Dependents','PhoneService', 'MultipleLines', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies','PaperlessBilling', 'Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes' : 1, 'No' : 0}, inplace = True)
```

```python
df1['gender'].replace({'Male': 0, 'Female' : 1}, inplace=True) 
```

```python
#for more than 2 categories
df2 = pd.get_dummies(data = df1, columns=['InternetService', 'Contract', 'PaymentMethod'])
df2.columns
```

```python
df2.sample(2)
```

```python
col_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler  = MinMaxScaler()
df2[col_scale] = scaler.fit_transform(df2[col_scale])
```

```python
x = df2.drop('Churn', axis= 'columns')
y = df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 5)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense((20), input_shape = (26,), activation = 'relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam',
             loss = 'binary_crossentropy',
              metrics = ['accuracy']
             )
model.fit(x_train, y_train, epochs=100)
```

```python
model.evaluate(x_test, y_test)
```

```python
yp=model.predict(x_test)
yp[:5]
```

```python
y_pred = []
for i in range(len(yp)):
    if yp[i]<0.5:
        y_pred.append(0)
    else:
        y_pred.append(1)
        
```

```python
y_pred[:5]
```

```python
#classification report
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

```python
#confusion matrix
import seaborn as sn
cm = tf.math.confusion_matrix(labels = y_test, predictions= y_pred)
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel("Truth")
```
