## My Dataset-  Music and Mental Health EDA

### Introduction

Music therapy, or MT, is the use of music to improve an individual's stress, mood, and overall mental health. MT is also recognized as an evidence-based practice, using music as a catalyst for "happy" hormones such as oxytocin.

However, MT employs a wide range of different genres, varying from one organization to the next.

The MxMH dataset aims to identify what, if any, correlations exist between an individual's music taste and their self-reported mental health. Ideally, these findings could contribute to a more informed application of MT or simply provide interesting sights about the mind.

The dataset is collected from a survey conducted to understand the realtionship between music and mental health. The respondents of the survey were not restricted by age or location.The form was relatively brief so that respondents would be more likely finish the survey. "Harder" questions (such as BPM) were left optional for the same reason.

### Domain Knowledge

Music has a profound impact on mental health and well-being. It can be a powerful tool for promoting emotional, psychological, and cognitive health. Here are some key ways in which music influences mental health:
1. Emotional Regulation: Music has the ability to evoke a wide range of emotions, from joy and excitement to sadness and nostalgia.  eople often use music to regulate their emotions, seeking out specific songs or genres that match their current mood or help them shift their emotional state.

2. Stress Reduction: Listening to calming music can lower stress levels and reduce the production of stress hormones like cortisol. Music with a slow tempo and soothing melodies can induce relaxation and alleviate anxiety.

3. Enhanced Cognitive Function: Music can stimulate cognitive functions such as memory, attention, and problem-solving. For individuals with neurodegenerative conditions like Alzheimer's disease, music can evoke memories and improve cognitive functioning.

4. Relaxation and Sleep: Calming music is frequently used to improve sleep quality. Listening to soothing music before bedtime can promote relaxation and help individuals fall asleep more easily.

5. Self-Care and Coping Mechanism: Many people turn to music as a form of self-care and a coping mechanism during difficult times. It can provide comfort, a sense of companionship, and a means of processing complex emotions.

The knowlege that will be required to understand this dataset will be:
1. Music Therapy Expertise: A solid foundation in music therapy principles and techniques is essential. This involves understanding how music is used to address emotional, cognitive, and social needs in therapeutic settings.

2. Psychological Understanding: A grasp of basic psychology concepts is crucial. This includes knowledge of emotions, cognition, and the psychological impact of music on mood, memory, and behavior. Familiarity with mental health conditions like anxiety and depression is also important.

3. Neuroscientific Insight: An understanding of the neuroscience of music is key. This encompasses knowledge about how music affects the brain, including its impact on neural pathways, neurotransmitters, and brain structures.

4. Assessment and Evaluation Skills: Competence in assessment tools and methodologies used in mental health is essential. This includes the ability to measure and evaluate changes in mental health outcomes resulting from music therapy interventions.

```python
#Q10 Are there any correlations between variables? Calculate correlations
#using one hot encoding
data_encoded = pd.get_dummies(data, columns=['While working', 'Instrumentalist', 'Composer', 'Exploratory', 'Foreign languages', ],drop_first=True)
#using label encoder for ordinal data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = ['Fav genre', 'Frequency [Classical]', 'Frequency [Country]', 'Frequency [EDM]',
       'Frequency [Folk]', 'Frequency [Gospel]', 'Frequency [Hip hop]',
       'Frequency [Jazz]', 'Frequency [K pop]', 'Frequency [Latin]',
       'Frequency [Lofi]', 'Frequency [Metal]', 'Frequency [Pop]',
       'Frequency [R&B]', 'Frequency [Rap]', 'Frequency [Rock]',
       'Frequency [Video game music]','Music effects', 'Primary streaming service']
for col in cols:
    data_encoded[col]= le.fit_transform(data[col])
data_encoded.head()

#finding correlation
correlation = data_encoded.corr()
plt.figure(figsize=(25,25))
sns.heatmap(correlation)
plt.show()
```

```python
#11 How do the age groups in the data set differ in terms of music preferences and mental health indicators?

#creating groups
data_pr = data
bins = [0, 18, 25, 35, 45, 55, 65, float("inf")]
labels = ["Under 18", "18-24", "25-34", "35-44", "45-54", "55-64", "65 and over"]

data_pr['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels)

age_group_data = data.groupby('Age Group').agg({
    'Hours per day': 'mean',
    'Anxiety': 'mean',
    'Depression': 'mean',
    'Insomnia': 'mean',
    'OCD': 'mean'
}).reset_index()
age_group_data
```

```python
fig, axes = plt.subplots(2,2, figsize=(12, 8))
indicators = ['Hours per day', 'Anxiety', 'Depression', 'Insomnia']

for i, indicator in enumerate(indicators):
    ax = axes[i // 2, i % 2]
    ax.bar(age_group_data['Age Group'], age_group_data[indicator])
    ax.set_title(f'Mean {indicator} by Age Group')
    ax.set_xlabel('Age Group')
    ax.set_ylabel(indicator)

plt.tight_layout()
plt.show()
```

Q12 What is the classification of the objective variable (if any)? Are there class imbalances, and if so, how important are they?
<br>There is class imbalance as there is difference in target data of Improve, No effect, Worsen

```python
#13 Are there significant differences in mental health symptoms based on the primary service used by individuals?
from scipy.stats import f_oneway

grouped_data = data.groupby('Primary streaming service')

state = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
service = ['Spotify', 'YouTube Music', 'I do not use a streaming service.',
       'Apple Music', 'Other streaming service', 'Pandora']
#Using ANOVA
for col in state:
    f_statistic, p_value = f_oneway(
            grouped_data.get_group('Spotify')[col],
            grouped_data.get_group('YouTube Music')[col],
            grouped_data.get_group('Apple Music')[col],
            grouped_data.get_group('Pandora')[col],
            grouped_data.get_group('I do not use a streaming service.')[col],
            grouped_data.get_group('Other streaming service')[col]
    )

    # Check significance
    alpha = 0.05
    if p_value < alpha:
        print(f"There are significant differences in {col} based on streaming service.")
    else:
        print(f"No significant differences in {col} based on streaming service.")
```

```python
#14 Do individuals who identify as instrumentalists or musicians have a different pattern of psychiatric symptoms compared to those who do not?
from scipy.stats import ttest_ind

for col in state:
    group1 = data_encoded[data_encoded['Instrumentalist_Yes'] == 1][col]
    group2 = data_encoded[data_encoded['Instrumentalist_Yes'] == 0][col]
    
    #using t test
    t_statistic, p_value = ttest_ind(group1, group2)

    # Check significance
    alpha = 0.05
    if p_value < alpha:
        print(f"There are significant differences in {col} between instrumentalistsand non-instrumentalists.")
    else:
        print(f"No significant differences in {col} between instrumentalists and non-instrumentalists.")
```

```python
#15. Are there any linguistic patterns in the data set, such as foreign-speaking individuals with different music preferences or mental health outcomes?
data_grouped = data.groupby('Foreign languages')

music_preferences = data_grouped['Fav genre'].value_counts()
mental_health= data_grouped[['Anxiety', 'Depression', 'Insomnia', 'OCD']].mean()

music_preferences.unstack().plot(kind='bar', stacked=True)
plt.title('Favorite Genre by Language Preference')
plt.xlabel('Language Preference')
plt.ylabel('Count')
plt.legend(title='Favorite Genre')
plt.show()

print(mental_health)
```

```python
#16 Is there a relationship between music tempo (BPM) and self-reported mental health symptoms such as anxiety and depression?

from scipy.stats import pearsonr
#calculating correlation coefficient
#anxiety
correlation_anxiety, p_value_anxiety = pearsonr(data['BPM'], data['Anxiety'])
#Depression
correlation_depression, p_value_depression = pearsonr(data['BPM'], data['Depression'])

print(f'Correlation between BPM and Anxiety: {correlation_anxiety:.2f}')
print(f'Correlation between BPM and Depression: {correlation_depression:.2f}')
```

```python
#17 Are there patterns of music preference and mental health symptoms based on frequency of exposure to specific genres?
genre_freq_groups = data.groupby('Fav genre')
most_popular_genre = genre_freq_groups['Fav genre'].value_counts().groupby(level=0).nlargest(1)
most_popular_genre
```

```python
#18 Do individuals who listen to video game music show any different traits or psychiatric outcomes compared to individuals who do not?
data_pr=data
data_pr['ListensToVideoGameMusic'] = data_pr['Frequency [Video game music]'].apply(lambda x: 1 if x in ['Sometimes', 'Very frequently'] else 0)

for col in state:
    group_listen = data_pr[data_pr['ListensToVideoGameMusic'] == 1][col]
    group_not_listen = data_pr[data_pr['ListensToVideoGameMusic'] == 0][col]

    t_statistic, p_value = ttest_ind(group_listen, group_not_listen)

    # Check significance
    alpha = 0.05
    if p_value < alpha:
        print(f"There are significant differences in {col} between individuals who listen to video game music and those who do not.")
    else:
        print(f"No significant differences in {col} between the two groups.")
```

```python
#19 Are there noticeable changes in psychiatric symptoms compared to individuals who report listening to music while working?
data_pr=data
data_pr['Listen'] =data['While working'].apply(lambda x: 1 if x == 'Yes' else 0)

for col in state:
    group_listen = data_pr[data_pr['Listen'] == 1][col]
    group_not_listen = data_pr[data_pr['Listen'] == 0][col]

    t_statistic, p_value = ttest_ind(group_listen, group_not_listen)

    # Check significance
    if p_value < alpha:
        print(f"There are significant differences in {col} between individuals who listen to music while working and those who do not.")
    else:
        print(f"No significant differences in {col} between the two groups.")
```
